```python
import dask.dataframe as dd
import dask.bag as db
import os
```

```python
df=dd.read_csv(os.path.join('air_traffic_data.csv')).compute()
df.head()
```

```python
df.info()
```

```python
a=df.groupby('Operating Airline IATA Code').count()
company=list(a.index)
print("hay", len(company), "compañias aereas")
```

```python
b=df.groupby('Operating Airline IATA Code')['Passenger Count'].mean()
print(b)
```

```python
df1=df.copy()
c= df.groupby('GEO Region')['Passenger Count'].max()
l=list(c)
l1=list(c.index)

borrar=[]

for i in range(len(df['Passenger Count'])):
    g=l1.index(df['GEO Region'][i])
    if df['Passenger Count'][i] != l[g]:
        borrar.append(i)

df=df.drop(borrar)
```

```python
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df.to_csv('Maximodepasajeros.csv')
b.to_csv('Mediadepasajeros.csv')
```
